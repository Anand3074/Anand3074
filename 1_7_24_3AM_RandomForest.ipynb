```python
def compute_recent_5_race_stats(group):
    group['Recent_5_Wins'] = group['Won'].rolling(window=5, min_periods=1).sum()  # Calculate rolling sum of wins
    group['Recent_5_Top_3'] = (group['Position'] <= 3).rolling(window=5, min_periods=1).sum()  # Calculate rolling sum of top 3 finishes

    # Calculate racing incidents in last 5 races
    group['Recent_5_Incidents'] = group['Racing_Incident'].rolling(window=5, min_periods=1).sum()

    # Calculate if racing incident in the last race
    group['Incident_Last_Race'] = group['Racing_Incident'].shift(1)

    return group
```

```python
data = data.groupby('Horse').apply(compute_recent_5_race_stats)
```

```python
def compute_speed_rating(row):
    if row['Time'] > 0:
        base_speed_rating = row['Dist'] / row['Time']
        base_speed_rating = convert_speed_rating(base_speed_rating)

        # Adjustment based on AvgWt - Wt
        weight_diff = row['Avg_Wt'] - row['Wt']
        base_speed_rating += weight_diff * 4

        # Adjustment based on jockey rank
        if row['Jockey_Rank'] == 1:
            base_speed_rating += 12
        elif row['Jockey_Rank'] == 2:
            base_speed_rating += 10
        elif row['Jockey_Rank'] == 4:
            base_speed_rating -= 4
        elif row['Jockey_Rank'] == 5:
            base_speed_rating -= 10

        # Adjustment based on allowance
        base_speed_rating += row['Allowance'] * -4

        # Adjustment based on racing incident in the last race
        if row['Incident_Last_Race'] == 1:
            base_speed_rating += 8

        # Adjustment based on racing incidents in the last 5 races
        if row['Recent_5_Incidents'] >= 3:
            base_speed_rating -= 4

        return base_speed_rating
    else:
        return None
```

```python
data['Speed_Rating'] = data.apply(compute_speed_rating, axis=1)
```

```python
def convert_odds(odds):
    try:
        if isinstance(odds, str) and '/' in odds:
            numerator, denominator = map(int, odds.split('/'))
            if denominator == 0:  # Check for zero denominator
                return None  # Handle division by zero
            return numerator / denominator + 1
        elif isinstance(odds, str) and odds.lower() == 'even':
            return 2.0
        elif isinstance(odds, (int, float)):
            return odds + 1
        else:
            return None  # Handle other cases where odds format is not recognized
    except ValueError:
        return None
```

```python
data.loc[:, 'Decimal_Odds'] = data['Odds'].apply(convert_odds)
```

```python
def fill_speedrating(group):
    mean_speed_rating = group['Speed_Rating'].mean(skipna=True)
    group['Speed_Rating'] = group['Speed_Rating'].fillna(mean_speed_rating)
    return group
```
