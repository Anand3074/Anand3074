```python
def compute_recent_5_race_stats(group):
    group['Recent_Wins'] = group['Won'].rolling(window=5, min_periods=1).sum()
    group['Recent_Top_3'] = (group['Position'] <= 3).rolling(window=5, min_periods=1).sum()
    group['Recent_Incidents'] = group['Racing_Incident'].rolling(window=5, min_periods=1).sum()
    group['Incident_Last_Race'] = group['Racing_Incident'].shift(1)
    return group
```

```python
def filter_trackwork_dates(row):
    race_date = row['Date']
    trackworks_within_window = merged_df[(merged_df['Horse'] == row['Horse']) &
                                         (merged_df['Trackwork_Date'] <= race_date) &
                                         (merged_df['Trackwork_Date'] >= race_date - pd.Timedelta(days=15))]
    if not trackworks_within_window.empty:
        # Select the most recent trackwork date
        most_recent_trackwork = trackworks_within_window.loc[trackworks_within_window['Trackwork_Date'].idxmax()]
        return most_recent_trackwork
    else:
        return pd.Series([pd.NaT] * len(merged_df.columns), index=merged_df.columns)
```

```python
def get_recent_trackwork(row, trackwork_df):
    horse = row['Horse']
    race_date = row['Date']
    # Filter trackwork within the 15-day window before the race date
    valid_trackworks = trackwork_df[(trackwork_df['Horse'] == horse) &
                                    (trackwork_df['Trackwork_Date'] <= race_date) &
                                    (trackwork_df['Trackwork_Date'] >= race_date - pd.Timedelta(days=15))]
    if not valid_trackworks.empty:
        # Get the most recent trackwork
        most_recent = valid_trackworks.loc[valid_trackworks['Trackwork_Date'].idxmax()]
        return most_recent
    else:
        return pd.Series([pd.NaT] * len(trackwork_df.columns), index=trackwork_df.columns)
```
