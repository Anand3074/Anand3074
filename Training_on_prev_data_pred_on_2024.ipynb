```python
data_path = '/content/drive/MyDrive/data_final.csv'
dataset_1 = pd.read_csv(data_path)
```

```python
dataset_1['Date'] = pd.to_datetime(dataset_1['Date'])
dataset_1['Year'] = dataset_1['Date'].dt.year
```

```python
dataset_1
```

```python
df_2024 = dataset_1[dataset_1['Year'] == 2024]
```

```python
df_2024
```

```python
dataset_1 = dataset_1[dataset_1['Year'] != 2024]
```

```python
dataset_1
```

```python
features = [
    'Wt', 'Dist', 'Class',
    'Decimal_Odds', 'Top_3','Win_PCT',
    'Recent_Wins', 'Recent_Top_3',
    'Predicted_Speed_Rating',
]
```

```python
target = 'Position'
```

```python
X = dataset_1[features + ['RaceID']]
y = dataset_1[target]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
train_race_ids = dataset_1.loc[X_train.index, 'RaceID']
train_race_ids
```

```python
val_race_ids = dataset_1.loc[X_test.index, 'RaceID']
val_race_ids
```

```python
group_train = X_train.groupby('RaceID').size().to_list()
group_test = X_test.groupby('RaceID').size().to_list()
```

```python
X_test_1 = X_test.copy()
```

```python
X_train = X_train.drop(columns=['RaceID'])
X_test = X_test.drop(columns=['RaceID'])
```

```python
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import ndcg_score
```

```python
train_data = lgb.Dataset(X_train, label=y_train, group=group_train)
```

```python
test_data = lgb.Dataset(X_test, label=y_test, group=group_test)
```

```python
params = {
    'objective': 'lambdarank',  # Objective for ranking
    'metric': 'ndcg',           # Metric for ranking
    'boosting_type': 'gbdt',
    'ndcg_at': [1, 3, 5],
    'learning_rate': 0.05,
    'num_leaves': 31,
    'feature_fraction': 0.8,
    'bagging_fraction': 0.8,
    'min_data_in_leaf': 20,
    'bagging_freq': 5,
    'lambda_l1': 0.1,  # Example regularization strength
    'verbose': 1,
    'label_gain': list(range(1, y_train.max() + 2))

}
```

```python
X_train
```

```python
early_stopping = lgb.early_stopping(stopping_rounds=10)
```

```python
train_data
```

```python
dataset_1
```

```python
num_rounds = 1000
bst= lgb.train(params,
                  train_data,
                  num_boost_round=num_rounds,
                  valid_sets=[train_data],  # Only need to validate on train_data for lambdarank
                  callbacks=[early_stopping],  # Stop if no improvement in 50 rounds
                )
```

```python
y_pred = bst.predict(X_test, num_iteration=bst.best_iteration)
```

```python
ndcg = ndcg_score([y_test], [y_pred])
print(f'NDCG Score: {ndcg}')
```
