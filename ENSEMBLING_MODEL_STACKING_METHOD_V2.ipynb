```python
# Ensure the dataset is sorted by date
dataset = dataset.sort_values(by=['Date'])

# Initialize columns for trainer and jockey win rates and podium counts
dataset['Trainer_Win_Rate'] = 0.0
dataset['Trainer_podium_count'] = 0
dataset['Jockey_Win_Rate'] = 0.0
dataset['Jockey_podium_count'] = 0

# Initialize dictionaries to keep track of wins and top 3 finishes for trainers and jockeys
trainer_win_counts = {}
trainer_podium_count_counts = {}
trainer_race_counts = {}

jockey_win_counts = {}
jockey_podium_count_counts = {}
jockey_race_counts = {}

# Iterate over the dataframe row by row
for index, row in dataset.iterrows():
    trainer = row['Trainer']
    jockey = row['Jockey']
    position = row['Position']

    # Initialize trainer statistics if not present
    if trainer not in trainer_win_counts:
        trainer_win_counts[trainer] = 0
        trainer_podium_count_counts[trainer] = 0
        trainer_race_counts[trainer] = 0

    # Initialize jockey statistics if not present
    if jockey not in jockey_win_counts:
        jockey_win_counts[jockey] = 0
        jockey_podium_count_counts[jockey] = 0
        jockey_race_counts[jockey] = 0

    # Calculate Win_Rate and podium_count up to this race for trainer
    if trainer_race_counts[trainer] > 0:
        dataset.at[index, 'Trainer_Win_Rate'] = trainer_win_counts[trainer] / trainer_race_counts[trainer]
        dataset.at[index, 'Trainer_podium_count'] = trainer_podium_count_counts[trainer]

    # Calculate Win_Rate and podium_count up to this race for jockey
    if jockey_race_counts[jockey] > 0:
        dataset.at[index, 'Jockey_Win_Rate'] = jockey_win_counts[jockey] / jockey_race_counts[jockey]
        dataset.at[index, 'Jockey_podium_count'] = jockey_podium_count_counts[jockey]

    # Update counts for trainer
    trainer_race_counts[trainer] += 1
    if position == 1:
        trainer_win_counts[trainer] += 1
    if position <= 3:
        trainer_podium_count_counts[trainer] += 1

    # Update counts for jockey
    jockey_race_counts[jockey] += 1
    if position == 1:
        jockey_win_counts[jockey] += 1
    if position <= 3:
        jockey_podium_count_counts[jockey] += 1

# Now the dataset contains the historical win rates and podium counts for trainers and jockeys
```

```python
import lightgbm as lgb
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold, train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)
```

```python
for train_index, valid_index in kf.split(train_data):
    X_train, X_valid = train_data.iloc[train_index], train_data.iloc[valid_index]
    y_train, y_valid = train_data['Position'].iloc[train_index], train_data['Position'].iloc[valid_index]

    # Train Model 1 (without odds)
    model1 = lgb.LGBMRegressor()
    model1.fit(X_train[features_without_odds], y_train)
    oof_predictions_model1[valid_index] = model1.predict(X_valid[features_without_odds])
    test_predictions_model1 += model1.predict(test_data[features_without_odds]) / kf.n_splits

    # Train Model 2 (with odds)
    model2 = lgb.LGBMRegressor()
    model2.fit(X_train[features_with_odds], y_train)
    oof_predictions_model2[valid_index] = model2.predict(X_valid[features_with_odds])
    test_predictions_model2 += model2.predict(test_data[features_with_odds]) / kf.n_splits
```

```python
meta_features = pd.DataFrame({
    'model1_pred': oof_predictions_model1,
    'model2_pred': oof_predictions_model2
})
```

```python
meta_model = LinearRegression()
meta_model.fit(meta_features, train_data['Position'])
```

```python
meta_test_features = pd.DataFrame({
    'model1_pred': test_predictions_model1,
    'model2_pred': test_predictions_model2
})
```

```python
final_predictions = meta_model.predict(meta_test_features)
```

```python
mse = mean_squared_error(test_data['Position'], final_predictions)
print(f'Mean Squared Error of the stacked model: {mse}')
```

```python
def predict_speed_rating_1(row, ):
    # Initialize predicted_rating based on availability and value of Recent_Avg_Speed_Rating
    predicted_rating = row['Avg_Speed_Rating']  # Use Recent_Avg_Speed_Rating as initial prediction

    # Adjust for distance change
    dist_diff = row['New_Dist'] - row['Dist']
    predicted_rating += (dist_diff // 100) * -10

    # Adjust for weight change only if prev_row is available
    wt_diff = row['New_Wt'] - row['Wt']
    # - row['Allowance']
    predicted_rating += wt_diff * -4

    # Adjust for jockey rank change
    if row['Jockey_Rank'] == 1:
        predicted_rating += 12
    elif row['Jockey_Rank'] == 2:
        predicted_rating += 10
    elif row['Jockey_Rank'] == 4:
        predicted_rating -= 4
    elif row['Jockey_Rank'] == 5:
        predicted_rating -= 10

    if row['Trainer_Ranking'] == 1:
        predicted_rating += 6
    elif row['Trainer_Ranking'] == 2:
        predicted_rating += 3
    elif row['Trainer_Ranking'] == 4:
        predicted_rating -= 3
    elif row['Trainer_Ranking'] == 5:
        predicted_rating -= 6

    class_diff = row['Class'] - row['Highest_Class']
    if class_diff == -1:  # Current class is higher than previous (dropped in class)
        predicted_rating +=  class_diff * 10
    elif class_diff == 1:  # Current class is lower than previous (moved up in class)
        predicted_rating -= class_diff * 10

    # Adjustment based on allowance only if prev_row is available
        predicted_rating += row['Allowance'] * -4

    # Adjustment based on racing incidents in the last 5 races
    # if row['Recent_Incidents'] >= 3:
    #     predicted_rating -= 4

    # Additional adjustments based on specific conditions (uncomment and modify as needed)
    # if row['Incident_Last_Race'] == 1:
    #     predicted_rating += 12

    if row['Sh'] == 'A':
        predicted_rating += 15
    # elif row['Sh'] == 'S':
    #     predicted_rating -= 15

    if row['Base_venueid'] == row['venueid']:
        predicted_rating += 8

    if row['Base_venueid'] != row['venueid']:
        predicted_rating -= 20

    if row['Change'] == 0:
        predicted_rating -= 8

    predicted_rating += row['Change'] * -2

    return predicted_rating
```
