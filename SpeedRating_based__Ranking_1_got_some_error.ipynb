```python
def convert_speed_rating(speed_rating):
    divided_value = speed_rating / 10
    decimal_part = divided_value - int(divided_value)
    speed_rating = round(decimal_part * 1000)
    return speed_rating
```

```python
def compute_speed_rating(row):
    if row['Time'] > 0:
        base_speed_rating = row['Dist'] / row['Time']
        base_speed_rating = convert_speed_rating(base_speed_rating)

        # Adjustment based on jockey rank
        if row['Jockey_Rank'] == 1:
            base_speed_rating += 12
        elif row['Jockey_Rank'] == 2:
            base_speed_rating += 10
        elif row['Jockey_Rank'] == 4:
            base_speed_rating -= 4
        elif row['Jockey_Rank'] == 5:
            base_speed_rating -= 10

        # Adjustment based on allowance
        base_speed_rating += row['Allowance'] * -4

        # Adjustment based on racing incident in the last race
        if row['Incident_Last_Race'] == 1:
            base_speed_rating += 12

        # Adjustment based on racing incidents in the last 5 races
        if row['Recent_Incidents'] >= 3:
            base_speed_rating -= 4

        # Adjustment based on Sh column
        if row['Sh'] == 'A':
            base_speed_rating += 15
        elif row['Sh'] == 'S':
            base_speed_rating -= 15

        return base_speed_rating
    else:
        return None
```

```python
def predict_speed_rating(row, prev_row):
    if pd.isna(prev_row['Speed_Rating']):
        return None

    predicted_rating = prev_row['Speed_Rating']

    # Adjust for distance change
    dist_diff = row['Dist'] - prev_row['Dist']
    predicted_rating += (dist_diff // 100) * -10

    # Adjust for weight change
    wt_diff = row['Wt'] - prev_row['Wt']
    predicted_rating += wt_diff * -4

    # Adjust for jockey rank change
    if row['Jockey_Rank'] == 1:
            predicted_rating += 12
    elif row['Jockey_Rank'] == 2:
            predicted_rating += 10
    elif row['Jockey_Rank'] == 4:
            predicted_rating -= 4
    elif row['Jockey_Rank'] == 5:
            predicted_rating -= 10

        # Adjustment based on allowance
    predicted_rating += row['Allowance'] * -4
    if prev_row['Incident_Last_Race'] == 1:
            predicted_rating += 12

        # Adjustment based on racing incidents in the last 5 races
    if row['Recent_Incidents'] >= 3:
            predicted_rating -= 4

    if row['Sh'] == 'A':
            predicted_rating += 15
    elif row['Sh'] == 'S':
            predicted_rating -= 15
    return predicted_rating
```
